```python
chat_history = ChatMessageHistory()
```

```python
context = """
The user is looking for a to buy or rent a house or an apartment. The goal is to ask questions that are specific and tailored based on the user's previous answers and questions to cover variety of aspects of apartments in our database so we can later recommend an apartment to them. 
"""
template = """
    You are a helpful assistant collecting buyer preferences for real estate properties. Your goal is to ask a series of thoughtful, tailored questions to understand the user's preferences. 
    
    Guidelines:
    1. Use the context below to guide the types of questions you ask.
    2. Tailor each subsequent question based on the user's previous answers. Although, feel free to ask different questions to cover as many aspects of the apartment and location. 
    3. Aim to gather information about property size, layout, amenities, location preferences, neighborhood type, and transportation needs.
    4. Ensure the questions encourage detailed and specific responses.

    Context:
    {context}
    
    Examples of questions you might ask. Start by asking only ONE question and wait to get a response. 
    
    - How big do you want your house to be?
    - What would be cozy living room, can you give me an example? Like a fireplace? 
    - I see, you want amenities. Which amenities would you like? A pool, club room? 
    - Which city do you want you next house to be? 
    - What is your price range? 
    - Are you looking to rent or buy? 
        
    """

prompt = PromptTemplate(template=template, input_variables=["context"])
chain = LLMChain(llm=llm, prompt=prompt)
```

### Option 1: Do it with the help of AI

```python
num_questions = 10
answers = []

previous_answer = ''

for i in range(num_questions):
    input_data = {"context": context}
    
    if previous_answer:
        input_data["context"] += f"\nPrevious Answer: {previous_answer}"

    question = chain.run(input_data).strip()
    print(f"\nQuestion {i+1}: {question}")

    user_answer = input("Your Answer: ")
    previous_answer = user_answer  # Tailor the next question based on this answer
    
    # Update the chat history with both the user's answer and the assistant's question
    chat_history.add_ai_message(question)
    chat_history.add_user_message(user_answer)
    
    
chat_history.add_ai_message("""Now recommend an apartment listing""")
```

## Parse chat history
We're going to parse the chat history and summarize it into a query that we can use to find similar documents in our database

```python
def parse_and_summarize_chat_history(chat_history):
    # Parse user and ai Q and A
    user_messages = [
        message.content for message in chat_history.messages if message.type == "human"
    ]
    ai_messages = [
        message.content for message in chat_history.messages if message.type == "ai"
    ]

    # Prompt to summarize it by LLM
    prompt = PromptTemplate(
        input_variables=["user_messages", "ai_messages"],
        template="""
        Based on the following user messages:
        {user_messages}

        And the assistant's responses:
        {ai_messages}

        Generate a concise summary of the user's preferences, constraints, and requirements for finding an apartment. This will server as the user preferences.  
        """,
    )

    # Combine the messages into the prompt
    summary = llm.generate(
        [prompt.format(
            user_messages=" ".join(user_messages),
            ai_messages=" ".join(ai_messages),
        )]
    )
    return summary

summary = parse_and_summarize_chat_history(chat_history)
query = summary.generations[0][0].text
```

## Find sematically similar documents

```python
# Let's find listing that are semantically similar 
similar_docs = db.similarity_search(query, k=5)
similar_docs
```

## We're going to augment the descriptions using the summarized query and return very personalized answers that are based on facts

```python
augmented_prompt = PromptTemplate(
        input_variables=["query", "description"],
        template="""
        A user is looking for an apartment based on the following query:
        {query}

        Here is a description of an apartment:
        {description}

        Rewrite the description to emphasize how well it matches the user's preferences. Do not make up facts, make sure the apartment details and the user preferences are true. 
        """
    )

# Augment each description based on the user query
chain = LLMChain(llm=llm, prompt=augmented_prompt)

augmented_results = []
for result in similar_docs:
    description = result.page_content
    augmented_description = chain.run({"query": query, "description": description})
    augmented_results.append(augmented_description)
```
